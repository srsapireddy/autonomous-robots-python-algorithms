```python
def plot(q1,q2,a1,a2,x,y):
    # list of all x points
    x0 = 0.0
    x1 = a1*np.cos(q1)
    x2 = a1*np.cos(q1)+a2*np.cos(q2+q1)
    x_list = [x0,x1,x2]
    # list of all y points
    y0 = 0.0
    y1 = a1*np.sin(q1)
    y2 = a1*np.sin(q1)+a2*np.sin(q2+q1)
    y_list = [y0,y1,y2]
    plt.scatter(x, y, label= "stars", color= "red",  marker= "*", s=200)
    plt.plot(x_list, y_list, color='green', linewidth = 3, marker='o', markerfacecolor='blue', markersize=6)
    plt.ylim(-(a1+a2),(a1+a2)) 
    plt.xlim(-(a1+a2),(a1+a2))
    plt.show()
    
```
